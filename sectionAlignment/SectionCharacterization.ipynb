```python
import json
import os
import gzip
import numpy as np



for lang in os.listdir('../gap/multiLanguageFromDumps/'): 
    print(lang)
    c = 0
    output = {}
    articlesDone = {} #avoid repeated (check why multiLanguageFromDumps has duplicated articles )
    with open('../gap/multiLanguageFromDumps/%s' % lang) as f: #this are huge files, jsonl, read line by line to avoid mem error
        for l in f:
            c+=1
            if c%1000000==0: print(c)
            tmp = json.loads(l)
            if tmp[0] not in articlesDone:
                articlesDone[tmp[0]] = ''    
                total_size = sum([data['size'] for sec,data in tmp[2].items()])
                total_links = sum([len(data['links']) for sec,data in tmp[2].items()])
                for sec,data in tmp[2].items():
                    size = data['size']
                    if size > 0: # if there is something in the sections, if nothing, ignore
                        pos = data['pos']
                        num_links = len(data['links'])
                        pos_inv = len(tmp[2]) - pos
                        pos_norm = pos / len(tmp[2])
                        link_dens = num_links / size
                        if total_links:
                            links_rel = num_links / total_links
                        else:
                            links_rel = 0
                        size_rel = size / total_size
                        output[sec] = output.get(sec,{'size':[],'pos':[],'pos_inv':[],
                                                      'pos_norm':[],'link_dens':[],'num_links':[],
                                                     'links_rel':[],'size_rel':[]})
                        output[sec]['size'].append(size)
                        output[sec]['pos'].append(pos)  
                        output[sec]['pos_inv'].append(pos_inv) #position from the bottom to the top
                        output[sec]['pos_norm'].append(pos_norm) # position relative to the total num of sec
                        output[sec]['link_dens'].append(link_dens) # ratio of link per charcacters
                        output[sec]['num_links'].append(num_links) # absolute number of links                 
                        output[sec]['links_rel'].append(links_rel) #ratio of the total links in the article in this section
                        output[sec]['size_rel'].append(size_rel) #relative size, considering all the sections
    summary = {}
    for section,data in output.items():
        summary[section] = {}
        for k,vals in data.items():
            summary[section]['%s_avg' % k] = np.mean(vals)
            summary[section]['%s_median' % k] = np.median(vals)
            summary[section]['%s_std' % k] = np.std(vals)
        summary[section]['TotalCount'] = len(vals)



    with open('../gap/SectionsCharacterization/%s' % lang ,'w') as f:
        json.dump(summary,f)

                                    
                
```
