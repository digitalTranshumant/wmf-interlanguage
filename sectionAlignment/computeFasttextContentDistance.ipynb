# Compute fasttext distances among sections content
Data coming from: parser_list_langs_current_FastText.py


```python
from scipy.spatial.distance import cosine
def cosine_dist(x):
    return 1 - cosine(x[0],x[1])
```

```python
import pandas,os,glob,numpy as np
vectors = sorted(glob.glob('multiLanguageFromDumpsFastextarticlesInSixLang/*'), key=os.path.getsize)

while vectors:
    lang1 = vectors.pop()
    print(lang1)
    lang1_code = lang1.split('_')[-1][:-5]
    df = pandas.read_json(lang1,lines=True).rename(columns={0:'q',1:lang1_code ,2:'vector'})
    df.drop_duplicates(subset=['q',lang1_code],inplace=True) #removing duplicates
    df = df[~df[lang1_code].str.contains('\[|{|"')] #remove noise
    df['vector'] = df.vector.apply(np.array)
    df.set_index('q',inplace=True)
    #df = df[df.index.isin(articleFilter)]
    for lang2 in vectors:
        lang2_code = lang2.split('_')[-1][:-5]
        print(lang2)
        df2 = pandas.read_json(lang2,lines=True).rename(columns={0:'q',1:lang2_code ,2:'vector'})
        df2.drop_duplicates(subset=['q',lang2_code],inplace=True) #removing duplicates
        df2 = df2[~df2[lang2_code].str.contains('\[|{|"')] #remove noise
        df2['vector'] = df2.vector.apply(np.array)       
        df2.set_index('q',inplace=True)
        #df2 = df2[df2.index.isin(articleFilter)]
        dfs = df.join(df2,how='inner',rsuffix='_2')
        pairs = zip(dfs['vector'],dfs['vector_2'])
        result = map(cosine_dist,pairs)
        #prepare output
        dfs.reset_index(inplace=True)
        dfs['fasttext_content_similarity'] = pandas.Series(result)
        del(dfs['vector'])
        del(dfs['vector_2'])
        dfs.dropna(inplace=True)
        dfs = dfs[[lang1_code ,lang2_code ,'fasttext_content_similarity']].groupby([lang1_code ,lang2_code]).agg(['mean','median']).reset_index()
        dfs.columns =['%s%s' % (x,y.upper()) for x,y in dfs.columns.tolist()]   #flatten index   
        dfs.to_pickle('fastTextContentSimilarities/%s_%s.p' % (lang1_code,lang2_code))
   
```
